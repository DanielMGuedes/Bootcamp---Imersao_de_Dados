```
# Importar Biblioteca
import pandas as pd 
# apontar para a fonte de dados
fonte = "https://github.com/alura-cursos/imersao-dados-2-2020/blob/ddff4f29df0d6470dd59910553899e549d0b5806/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true"

#Fazer a Leitura do Cabeçalho da Fonte
dados = pd.read_csv(fonte)
dados.head()
```

```
# mostrar os totais do arquivo: nesse casso 127380 - tuplas e 136 colunas
dados.shape
```

```
#Olhar coluna por coluna 
dados ["SG_UF_RESIDENCIA"]
```

```
# VER TODOS OS NOMES DAS COLUNAS
dados.columns.values
```

```
# Selecionando mais de uma Coluna: 
dados [['SG_UF_RESIDENCIA', 'Q025']]
```

```
# Trás a descrição dos elementos da coluna (1 amostra de cada)
dados['SG_UF_RESIDENCIA'].unique()
```

```
# o len faz a contagem de elementos./ Associado ao unique - fez a contagem de quantos 
#elementos diferentes
len(dados['SG_UF_RESIDENCIA'].unique())
```

```
# FAzer a contagem de toda a coluna por elementos(Estado):
dados['SG_UF_RESIDENCIA'].value_counts()
```

```
# FAzer a contagem por estados
dados['NU_IDADE'].value_counts()
```

```
#Mostrar a Amostra por estados e idade
dados[['NU_IDADE','SG_UF_RESIDENCIA']].value_counts()
```

```
# FAzer a contagem por idade e mostrar de forma ordenada (por valor absoluto)
dados['NU_IDADE'].value_counts().sort_index()
```

```
# UTILZIAR A CONTAGEM por idade para fazer um histograma 
# Histograma - Range e Frequencia 
dados['NU_IDADE'].hist()
```

```
# Histograma - Range e Frequencia - Com o controle da distribuição(Caixinhas)
dados['NU_IDADE'].hist(bins = 60, figsize = (10,8))
```

```
dados.columns.values
```

```
# Analisando a coluna 'IN_TREINEIRO'
dados['IN_TREINEIRO']
```

```
# Analisando a coluna 'IN_TREINEIRO' - Fazer uma query para buscar so os valores 1(True)
dados.query('IN_TREINEIRO == 1')
```

```
# Analisando a coluna 'IN_TREINEIRO' - Comparar para a coluna completa
dados.query('IN_TREINEIRO == 1')['IN_TREINEIRO']
```

```
# Relação entre idades e contagem dos treineiros por idade ** necessário ordenar pelo index da idade.. para ver..
dados.query('IN_TREINEIRO == 1')['NU_IDADE'].value_counts().sort_index()
```

```
# Olhar as Notas de Redação - HISTOGRAMA
dados['NU_NOTA_REDACAO'].hist(bins = 20, figsize=(8,6))
```

```
# Olhar as Notas de LINGUAGENS E CÓDIGO - HISTOGRAMA
dados['NU_NOTA_LC'].hist(bins = 20, figsize=(8,6))
```

```
# calcular as médias
dados['NU_NOTA_REDACAO'].mean()
```

```
# calcular o Desvio Padrão
dados['NU_NOTA_REDACAO'].std()
```

```
# utilizar o (DESCRIBE DO PANDAS) para trazer várias análises automáticas
provas =['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_REDACAO']

# depois de criar a variavel 'provas' ele trouxe vairias análises:
#contagem(count) ; media(mean); Desvio Padrão(std); min; max; 1º,2ºe3ºQuartil
dados[provas].describe()
```

```
# Analisar uma amostra a QUANTIDADE '''.quantile()'' de itens dentro de uma parametro '(0.9)'
# O resultado é o percentual de 10% das pessoas que tiraram nota mais alta de 596,3
dados['NU_NOTA_LC'].quantile(0.9)
```

```
# Agora vamos griar Grafico (boxPlot) para ver a distribuição da média da coluna
# selecionada distribuida em relação aos totais / 
#Função: plot.box()/ PArametro (grid = True)-Faz aparecer a grade e figsize(x,x) define o tamanho do box
dados['NU_NOTA_LC'].plot.box(grid = True, figsize=(8,6))

#
```

A Utilização de box.plot é melhor aproveitada para comparação de mais de uma variavél onde vai ser possivel ver graficamente o ponto médio entre diversos parametros
#

```
#Criar o gráfico comparando varias notas atraves de boxplot
dados[provas].plot.box(grid = True, figsize=(10,8))
```

É possível ver através da comparação de Notas que algumas ficaram mais concentradas (Quadrados maiores e menores) e a dispersão dos outliers.

Aplicação do conceito de 'Curva Normal' https://pt.wikipedia.org/wiki/Distribui%C3%A7%C3%A3o_normal  .

```
# Descobrir de quais estados são os inscritos com 13 anos
dados.query('NU_IDADE == 13')
```
