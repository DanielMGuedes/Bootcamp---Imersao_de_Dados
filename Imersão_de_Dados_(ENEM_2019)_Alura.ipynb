> Proximo Exercicio vamos plotar (boxplot) comparando a SOMA DE TODAS AS NOTAS e a RENDA

```
# Somando todas as Notas:
dados[provas].sum()
```

na execução acima é realizada a soma por coluna... e o que queremos é a soma das notas por linha.. para ver o TOTAL DA NOTA POR ALUNO


```
dados[provas].sum(axis=1)

dados['NU_NOTA_TOTAL'] = dados[provas].sum(axis=1) # Criamos uma nova coluna com as notas somadas por linha
dados.head()
```

```
# Agora criamos outro boxplot por total da Nota: 
plt.figure(figsize=(10,8)) # Aumentar tamanho
sns.boxplot(x='Q006', y ='NU_NOTA_TOTAL', data = dados, order = renda_ordenada) # Box Plot Ordenado
plt.title ('BoxPlot das notas de matemática em relação a Renda')

```

Verifica-se da análsie do Box acima que os valores de Notas 0 na popolação com baixa renda. enquanto que as maiores rendas concentram as Notas totais Maiores


```
# Fazendo histograma pra explorar melhor os dados verificados
sns.displot(dados, x='NU_NOTA_TOTAL')
```

Da analise do histograma só é possivl concluir que existe uma amostra muit grande da população que teve nota 0. Vamos explorar mais essa situação:


```
provas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_REDACAO']
```

```
provas.append('NU_NOTA_TOTAL')
dados[provas].query('NU_NOTA_TOTAL == 0')
```

Com a analise acima é possível verificar que as notas 0 são referentes as pessoas que não foram fazer as Provas.


```
# Criar uma nova variavel com as Notas sem Nota ZERO
dados_sem_notas_zero = dados.query('NU_NOTA_TOTAL != 0')
dados_sem_notas_zero.head()
```

```
plt.figure(figsize=(10,8)) # Aumentar tamanho
sns.boxplot(x='Q006', y ='NU_NOTA_TOTAL', data = dados_sem_notas_zero, order = renda_ordenada) # B
plt.title ('BoxPlot das notas total pela renda')
```

Incluir na análsie de a pessoa é reineiro ou não


```
plt.figure(figsize=(14,8)) # Aumentar tamanho
sns.boxplot(x='Q006', y ='NU_NOTA_TOTAL', data = dados_sem_notas_zero,
            hue ='IN_TREINEIRO',
            order = renda_ordenada) # Box Plot Ordenado
plt.title ('BoxPlot das notas total pela renda e se Treineiro')
```
